```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from numpy import genfromtxt
data_11= genfromtxt('../graphs/run_cufs_ae-tag-test_tloss.csv', delimiter=',')
data_12 = genfromtxt('../graphs/run_cufs_1215_183659-tag-test_tloss.csv', delimiter=',')
data_13 = genfromtxt('../graphs/run_celeba_1216_140817-tag-test_tloss.csv', delimiter=',')


data_21= genfromtxt('../graphs/run_cufs_ae-tag-tloss.csv', delimiter=',')
data_22 = genfromtxt('../graphs/run_cufs_1215_183659-tag-tloss.csv', delimiter=',')
data_23 = genfromtxt('../graphs/run_celeba_1216_140817-tag-tloss.csv', delimiter=',')
```

```python
plt.plot(data_11[1:,1], data_11[1:,2], 'b', data_12[1:,1],data_12[1:,2], 'r')
plt.plot(data_13[1:,1], data_13[1:,2], 'g')

plt.plot(data_21[1:,1], data_21[1:,2], 'b--', data_22[1:,1], data_22[1:,2],'r--')
plt.plot(data_23[1:,1], data_23[1:,2], 'g--')
plt.legend(['Test AE' ,'Test CUFS VAE','Test CELEBA VAE','AE' ,'CUFS VAE','CELEBA VAE'])
plt.xlabel('iterations')
plt.ylabel('Loss')
plt.savefig('../graphs/loss.eps')
```
